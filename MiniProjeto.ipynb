```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('faturamento_empresa.csv')

# Tratamento e Features
df['Data_Venda'] = pd.to_datetime(df['Data_Venda'])


# Forma de pagamento com mais transferências

gpby_payament = df.groupby('Forma_Pagamento')['Faturamento'].sum().sort_values(ascending=False)
print(f'Forma de pagamento que recebeu mais transferências: {gpby_payament.index[0]} | R${gpby_payament.iloc[0]}')


# Qual categoria gerou mais faturamento?

gpby_category = df.groupby('Categoria')['Faturamento'].sum().sort_values(ascending=False)
print(f'Categoria com mais faturamento: {gpby_category.index[0]}')


# Produtos para gráfico

gpby_product = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
#print(f'Top 3 Produtos mais Vendidos: \n- {gpby_product.iloc[0]} | {gpby_product.index[0]}\n- {gpby_product.iloc[1]} | {gpby_product.index[1]}\n- {gpby_product.iloc[2]} | {gpby_product.index[2]}')

# Cores para ficar crescente
colors = plt.cm.Greens(np.linspace(1, 0.5, len(gpby_product)))

plt.figure()
plt.bar(gpby_product.index, gpby_product.values, color=colors, label='Produto mais vendido')
plt.xlabel('Nome dos Produtos')
plt.ylabel('Quantidade Vendida')
plt.title('Ranking de Produtos mais Vendidos')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Verificar valores nulos
missing_data = df.isnull().sum().reset_index()
display(missing_data)

# Gráfico para mostrar valores nulos
missing_data.plot(kind='bar', color='red')
plt.title('Dados Faltantes por Coluna')
plt.ylabel('Quantidade de Dados Faltantes')
plt.show()
```
